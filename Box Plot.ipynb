#BOX plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mart = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Sample_data/supermarket_sales.xlsx')
```

```python
sns.boxplot(x='Total', data=mart)
```

# adding Width

```python
sns.boxplot(x='Total', data=mart, width=0.5)
```

# Changing to Y axis

```python
sns.boxplot(y='Total', data=mart, width=0.2)
```

# using x and y axis and color palette

```python
sns.boxplot(x='Payment', y='Total', data=mart, palette='Set3')
```

# adding hue 3 value use gender

```python
sns.boxplot(x='Payment', y='Total', hue='Gender', data=mart, palette='Set3')
```

# show means

```python
sns.boxplot(x='Payment', y='Total', hue='Gender', data=mart,showmeans=True, palette='Set3')
```
